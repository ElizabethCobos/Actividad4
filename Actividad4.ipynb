```python
#IMPORTAMOS PAQUETERÍAS
import pandas as pd
import numpy as np
import cufflinks as cf
import matplotlib.pyplot as plt
from IPython.display import display,HTML

cf.set_config_file(sharing='public',theme='white',offline=True) # write cf.getThemes() to find all themes available
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('COVID19Mexico.csv') 
data
```

```python
#Detección de valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data1=data.copy()
```

```python
#Sustituir valores nulos con promedio
data1["SECTOR"]=data1["SECTOR"].fillna(round(data1["SECTOR"].mean(),1))
data1
```

```python
#Corroboramos valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data2=data1.copy()
```

```python
#Sustituir valores nulos con mediana
data2["ENTIDAD_UM"]=data2["ENTIDAD_UM"].fillna(round(data2["ENTIDAD_UM"].median(),1))
data2
```

```python
#Corroboramos valores nulos
valores_nulos=data2.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data3=data2.copy()
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  "forward fill" ("ffill")
#Filtro por columnas
data3["PAIS_NACIONALIDAD"] =data3["PAIS_NACIONALIDAD"].fillna(method="ffill") 
data3
```

```python
#Corroboramos valores nulos
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data4=data3.copy()
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
#Filtro por columnas
data4["SEXO"] =data4["SEXO"].fillna(method="bfill") 
data4
```

```python
#Corroboramos valores nulos
valores_nulos=data4.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data5=data4.copy()
```

```python
#Sustituir valores nulos por un valor numérico en  concreto 
#Filtro por columnas
data5["EMBARAZO"]= data5["EMBARAZO"].fillna(2) 
data5
```

```python
#Corroboramos valores nulos
valores_nulos=data5.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data6=data5.copy()
```

```python
#Sustituir valores nulos por un valor en  concreto 
#Filtro por columnas
data6["MUNICIPIO_RES"]= data6["MUNICIPIO_RES"].fillna(95) 
data6
```

```python
#Corroboramos valores nulos
valores_nulos=data6.isnull().sum()
valores_nulos
```

```python
#Realizamos una copia del dataframe
data7=data6.copy()
```

```python
#Sustituir valores nulos con promedio
data7["EDAD"]=data7["EDAD"].fillna(round(data7["EDAD"].mean(),1))
data7
```

```python
#Corroboramos valores nulos
valores_nulos=data7.isnull().sum()
valores_nulos
```

```python
#Convertir DataFrame a CSV
data7.to_csv("COVID_sindatosnulos.csv")
```

#EN ESTA PARTE DEL CÓDIGO ANALIZAREMOS LOS OUTLIERS DE LA BASE DE DATOS "data_fil"

```python
#ORIGEN
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
y=data_fil["ORIGEN"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Outliers
outliers_iqr= data_fil[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= data_fil[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data_clean_iqr
```
